# psql 클라이언트

## 프롬프트

- 데이터베이스 이름 + 현재 상태를 나타내는 특수기호

- `dbname=#`: 슈퍼유저에서 새로운 명령 입력 대기
- `dbname=>`: 일반 사용자에서 새로운 명령 입력 대기
- `dbname->`: 명령이 세미콜론으로 끝나지 않음
- `dbname(>`: 괄호가 닫히지 않음
- `dbname'>`: 작은따옴표(')가 닫히지 않음
- `dbname">`: 큰따옴표(")가 닫히지 않음
- `dbname$>`: $$나 $tag$가 닫히지 않음
- `dbname/*>`: 주석이 닫히지 않음

## 메타 명령

- psql에서만 사용가능한 특수 명령
- `\`(백슬래시)로 시작하고 세미콜론으로 끝나지 않음

## 버퍼

- `;` 입력 후 엔터를 치기 전까지는 버퍼에 입력

### 버퍼 관련 메타 명령

- `\p`: 버퍼 내용 출력
- `\r`: 버퍼 리셋
