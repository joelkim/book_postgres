# 롤

## 롤

- 롤이 유일한 실체이며 유저와 그룹은 특정한 성질을 가지는 롤을 가리킨다
  - 유저: 로그인 권한을 가진 롤
  - 그룹: 다른 롤(유저)를 멤버로 포함하는 롤
- 관례상 하나의 롤이 유저의 역할과 그룹의 역할을 동시에 하지 않도록 관리한다.

### 관련 명령어

- 생성/수정/삭제
  - `CREATE ROLE`
  - `ALTER ROLE`
  - `DROP ROLE`

- 메타 명령
  - `du+`

- 시스템 카탈로그 조회
  - `SELECT * FROM pg_roles;`

- 현재 접속중인 롤 조회
  - `SELECT current_role;`

## 롤 생성

### 롤 생성 명령

- `CREATE` 명령으로 생성
  ```sql
  CREATE ROLE name [ [ WITH ] option [ ... ]]
  ```
- 한 번에 하나의 롤만 생성 가능 


### 롤 생성 옵션

- 옵션은 `WITH` 뒤에 오는 문자열
- 하나의 단어 혹은 복수의 단어로 이루어짐
- 추가 인수를 가지는 옵션도 있음
- 여러개의 옵션을 `WITH` 뒤에 한 번에 붙여서 사용할 수 있음
- 한 쌍으로 이루어진 옵션의 경우
  - `NO`가 붙지 않는 옵션은 기능을 부여
  - `NO`가 붙은 옵션은 기능을 제한(삭제)
  - 예시: `LOGIN`, `NOLOGIN`
  - 둘 중 하나가 디폴트값

- 로그인 및 인증 관련
  - `LOGIN`/`NOLOGIN`
  - `PASSWORD 문자열`
  - `PASSWORD NULL`
  - `CONNECTION LIMIT 숫자`
  - `VALID UNTIL 시간`

## 그룹

- 유저 롤을 그룹 롤에 추가하는 방법
  1. `CREATE` 명령으로 유저 롤 생성시 그룹 롤을 명시
  2. `GRANT` 명령으로 존재하는 유저 롤을 그룹 롤에 포함 

## 호스트 기반 인증

### `pg_hba.conf` 파일

- 모든 데이터베이스 연결 권한은 호스트 서버 내의 `pg_hba.conf` 파일에서 설정한다.
- hba 파일은 "Host Based Authentication"의 약자
